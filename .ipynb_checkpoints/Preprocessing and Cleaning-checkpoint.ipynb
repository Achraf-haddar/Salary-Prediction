```python
import pandas as pd
```

```python
df = pd.read_csv('df.csv')
print(df.shape)
df.head()
```

```python
# get salary estimate != -1
df = df[df["Salary Estimate"] != '-1']
print(df.shape)
```

```python
df['Salary Estimate'] = df['Salary Estimate'].apply(lambda x: x.split('(')[0])
df.head()
```

```python
df['Salary Estimate'] = df['Salary Estimate'].apply(lambda x: x.replace('K',''))
```
